## Function for the searching API

```python
def querycensys(query):
    API_URL = "https://censys.io/api/v1"
    res = requests.post(API_URL+"/search/ipv4", data=json.dumps(query),auth=(UID, SECRET))

    if res.status_code != 200:
        print "error occurred: %s" % res.json()["error"]
        sys.exit(1)
    
    return res
```

```python
results=querycensys({ "query":"location.country_code: NL"})
```
